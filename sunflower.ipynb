```python
import pandas as pd
import numpy as np
```

```python
def stad_scalers(values=[list]) -> list:
    return (values - np.mean(values)) / np.std(values)
```

```python
def min_max_scalar(values=[list]) -> list:
    return (values - min(values)) / (max(values) - min(values))
```

```python
a=[12,23,34,54,56,67,67]
stad_scalers(a)
print(np.mean(a))
print(np.std(a))
```

```python
stad_scalers(a)
```

```python
data = {
    'name':['sachin','virat','rohit','dhoni','rahul','surya','hardik','bahji','yuvraj','rishabh'],
    'runs':np.random.randint(3000,5000,10),
    'city':['mumbai','dilhi','mumbai','ranchi','mumbai','varodara','punjab','Jalandhar','Chandigarh','Roorkee'],
    'age':np.random.randint(26,45,10)
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['scaled runs'] =stad_scalers(df['runs'])
```

```python
df['normalize age'] = min_max_scalar(df['age'])
```
